```python
import boto3
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()

s3 = boto3.client(
    "s3",
    aws_access_key_id=os.getenv("AWS_ACCESS_KEY_ID"),
    aws_secret_access_key=os.getenv("AWS_SECRET_ACCESS_KEY"),
    region_name=os.getenv("AWS_REGION")
)

bucket = "proyecto5c-datos"
# Aseg√∫rate que la may√∫scula 'Dataset' coincida exactamente con tu bucket
key = "DataSet/yellow_tripdata_2016-01.csv" 

try:
    # Obtenemos el objeto (la conexi√≥n al archivo)
    obj = s3.get_object(Bucket=bucket, Key=key)
    
    # TRUCO DE BIG DATA: 
    # En lugar de leer todo, le decimos a Pandas que lea directamente del stream
    # y que PARE despu√©s de 5 filas (nrows=5).
    df = pd.read_csv(obj['Body'], nrows=5)
    
    print("‚úÖ Archivo le√≠do correctamente (solo primeras 5 filas).")
    print("Columnas encontradas:")
    print(df.columns.tolist())
    
    # Verificaci√≥n
    if 'tip_amount' in df.columns:
        print("\nüéâ La columna 'tip_amount' S√ç existe.")
    else:
        print("\n‚ö†Ô∏è No veo 'tip_amount'. Revisa la lista de columnas arriba.")

except Exception as e:
    print(f"Error: {e}")
```
