```python
from dotenv import load_dotenv
import os

load_dotenv()

aws_id = os.getenv("AWS_ACCESS_KEY_ID")
supabase_url = os.getenv("SUPABASE_URL")
supabase_key = os.getenv("SUPABASE_KEY")

print("ENV cargado correctamente:")
print("  - AWS_ACCESS_KEY_ID:", aws_id[:4] + "****" if aws_id else "NO DETECTADO")
print("  - SUPABASE_URL:", supabase_url if supabase_url else "NO DETECTADO")
print("  - SUPABASE_KEY:", supabase_key[:6] + "****" if supabase_key else "NO DETECTADO")
```

```python
import boto3

try:
    s3 = boto3.client("s3")
    buckets = s3.list_buckets()

    print("Conexión a AWS S3: OK")
    print("Buckets encontrados:")

    for b in buckets["Buckets"]:
        print("  -", b["Name"])

except Exception as e:
    print("Error al conectar a S3:", e)
```

```python
from supabase import create_client

try:
    supabase = create_client(os.getenv("SUPABASE_URL"), os.getenv("SUPABASE_KEY"))

    print("Conexión a Supabase: OK")
    print("Cliente inicializado correctamente:", supabase is not None)

except Exception as e:
    print("Error al conectar a Supabase:", e)

```
